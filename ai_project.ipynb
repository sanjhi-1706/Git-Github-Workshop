```python
def detect_accident(image_path, conf_threshold=0.5):
    results = yolo_model.predict(image_path, conf=conf_threshold)  # Run YOLO on image

    accident_detected = False  # Default: No accident

    for result in results:
        for box in result.boxes:
            label = result.names[int(box.cls[0])]  # Get class label
            confidence = box.conf[0]  # Get confidence score

            if label == "accident" and confidence > conf_threshold:
                accident_detected = True
                break  # Stop checking once an accident is found

    if accident_detected:
        print("⚠️ Accident Happened!")
    else:
        print("✅ No Accident Detected.")

# Test with an image
detect_accident("/content/runs/detect/train13/images_735-2-_jpg.rf.b40dc2559fa2b9598af5612acec1ce0a.jpg")
```

```python

def detect_accident_in_video(video_path, conf_threshold=0.2):
    cap = cv2.VideoCapture(video_path)  # Open video file
    accident_detected = False

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break  # Stop if video ends

        # Run YOLO detection on the frame
        results = yolo_model(frame)

        for result in results:
            for box in result.boxes:
                label = result.names[int(box.cls[0])]  # Get class label
                confidence = box.conf[0]  # Confidence score

                if label == "accident" and confidence > conf_threshold:
                    accident_detected = True  # Mark accident detected
                    cap.release()
                    cv2.destroyAllWindows()
                    print("⚠️ Accident Happened in Video!")
                    return  # Stop checking after detecting an accident

    cap.release()
    cv2.destroyAllWindows()
    print("✅ No Accident Detected in Video.")

# Test with a video
detect_accident_in_video("/content/segments-017.mp4")
```

```python
import requests

def get_current_location(api_token):
   try:
        response = requests.get(f"https://ipinfo.io/json?token={api_token}")
        data = response.json()
        loc = data['loc'].split(',')
        lat, lon = float(loc[0]), float(loc[1])
        print(f"📍 Location: {data['city']}, {data['region']}, {data['country']}")
        print(f"🌐 Coordinates: {lat}, {lon}")
        return lat, lon
   except Exception as e:
        print("❌ Error fetching location:", e)
        return None

# Example usage
get_current_location("6cc277dc69aff5")
```

```python
import requests



def get_nearby_hospital(lat, lon, api_key):
    url = (
        f"https://api.geoapify.com/v2/places"
        f"?categories=healthcare.hospital"
        f"&filter=circle:{lon},{lat},5000"
        f"&bias=proximity:{lon},{lat}"
        f"&limit=3"
        f"&apiKey={api_key}"
    )

    response = requests.get(url)
    hospitals = response.json().get("features", [])

    if hospitals:
        nearest = hospitals[0]["properties"]
        print("🏥 Nearest Hospital:", nearest["name"])
        print("📍 Address:", nearest["formatted"])
        return nearest
    else:
        print("❌ No hospitals found nearby.")
        return None



lat, lon = get_current_location()
if lat and lon:
    get_nearby_hospital(lat, lon, api_key="7b513e9863a0441296c99881c47d9aec")
```

```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

def send_accident_email(hospital_name, hospital_address, user_lat, user_lon, to_email):
    from_email = "youremail@gmail.com"
    app_password = "your_app_password"  # Use Gmail app password

    subject = "🚨 Accident Alert - Immediate Attention Needed"
    body = f"""
    An accident has been detected near the following location:

    📍 Latitude: {user_lat}
    📍 Longitude: {user_lon}

    Nearest Hospital: {hospital_name}
    Address: {hospital_address}

    Google Maps: https://www.google.com/maps/search/?api=1&query={user_lat},{user_lon}

    Please dispatch emergency help if needed.
    """

    msg = MIMEMultipart()
    msg["From"] = from_email
    msg["To"] = to_email
    msg["Subject"] = subject
    msg.attach(MIMEText(body, "plain"))

    try:
        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()
        server.login(from_email, app_password)
        server.sendmail(from_email, to_email, msg.as_string())
        server.quit()
        print("✅ Accident report sent via email.")
    except Exception as e:
        print("❌ Failed to send email:", e)

hospital = get_nearby_hospital(lat, lon, api_key="7b513e9863a0441296c99881c47d9aec")

if hospital:
    send_accident_email(
        hospital_name=hospital["name"],
        hospital_address=hospital["formatted"],
        user_lat=lat,
        user_lon=lon,
        to_email="hospital.alerts@example.com"  # can be your test email
    )
```
