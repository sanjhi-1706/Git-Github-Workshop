```python
from ultralytics import YOLO
model=YOLO("yolo11n.pt")
```

```python
from sahi import AutoDetectionModel

# Define the model_path variable
model_path = "yolo11n.pt"  # Replace with the actual path to your YOLO model if different

detection_model = AutoDetectionModel.from_pretrained(
    model_type="ultralytics",
    model_path=model_path,
    confidence_threshold=0.3,
    device="cpu",  # or 'cuda:0'
)
```

```python
from sahi.predict import get_sliced_prediction

# Update the image path
image_path = "/content/chidiyaghar.jpg"  # Assume the image is at the root of your Colab environment

result = get_sliced_prediction(
    image_path,  # Changed to image_path
    detection_model,
    slice_height=512,
    slice_width=512,
    overlap_height_ratio=0.2,
    overlap_width_ratio=0.2,
)

# Now you can access the predictions in the 'result' object
result.export_visuals(export_dir="sahi_out/")
```
